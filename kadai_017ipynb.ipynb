```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}さんは大人です。")
        else:
            print(f"{self.name}さんは大人ではありません。")

# 複数のインスタンスを生成
humans = [
    Human("太郎", 18),
    Human("花子", 22),
    Human("次郎", 30)
]

# リストの要素数分だけcheck_adultを呼び出す
for human in humans:
    human.check_adult()
```

```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}さんは大人です。")
        else:
            print(f"{self.name}さんは大人ではありません。")

# 複数のインスタンスを生成してリストに追加
humans = [
    Human("太郎", 18),
    Human("花子", 22),
    Human("次郎", 30),
    Human("三郎", 15)
]

# リストの要素数分だけ check_adult を呼び出す
for human in humans:
    human.check_adult()
```
