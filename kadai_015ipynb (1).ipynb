```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        # 属性の定義
        self.name = name
        self.age = age

    # 名前と年齢を出力するメソッド
    def printinfo(self):
        print("名前:", self.name)
        print("年齢:", self.age)


# Humanクラスのインスタンスを作成し、変数に代入
person = Human("Taro", 25)

# メソッドを呼び出して出力
person.printinfo()
```
