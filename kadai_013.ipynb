```python
# -*- coding: utf-8 -*-
"""
課題013: 消費税を計算する関数
"""

# 消費税を計算して返す関数
def calc_tax(price, tax_rate):
    total = price + (price * tax_rate / 100)
    return total

# 商品価格と税率
item_price = 110
tax_percent = 10

# 関数を呼び出して計算し、結果を表示
total_price = calc_tax(item_price, tax_percent)
print("消費税込みの金額:", total_price)
```
