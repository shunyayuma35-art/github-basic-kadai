```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        # 属性（変数のスコープ：インスタンス変数）
        self.name = name
        self.age = age

    # 成人かどうかをチェックするメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}（{self.age}歳）は大人です。")
        else:
            print(f"{self.name}（{self.age}歳）は大人ではありません。")

# ===== メイン処理部分 =====

def main():
    # Humanインスタンスを複数生成してリストに追加
    humans = [
        Human("太郎", 18),
        Human("花子", 22),
        Human("次郎", 20),
        Human("美咲", 15)
    ]

    # リストの要素数分だけ check_adult を呼び出す
    for human in humans:
        human.check_adult()

# 実行
if __name__ == "__main__":
    main()
```
