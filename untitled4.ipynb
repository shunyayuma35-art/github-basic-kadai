```python
# 台形の辺の長さを変数に代入
top = 10      # 上辺
bottom = 20   # 下辺
height = 5    # 高さ

# 台形の面積を計算して変数に代入
area = (top + bottom) * height / 2

# 数値と文字列を結合して出力
print(str(area) + "cm²")
```
