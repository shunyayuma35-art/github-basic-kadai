```python
# -*- coding: utf-8 -*-
"""
課題013: 消費税を計算する関数
"""

# 消費税を計算して返す関数
def calc_tax(price, tax_rate):
    """
    price: 商品の金額
    tax_rate: 消費税率（%）
    戻り値: 消費税込みの金額
    """
    total = price + (price * tax_rate / 100)
    return total

# 関数を呼び出して計算
item_price = 110   # 商品の金額
tax_percent = 10   # 消費税率（%）
```
