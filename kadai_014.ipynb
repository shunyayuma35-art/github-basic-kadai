```python
# =========================================
# 課題14：変数スコープを正しく使おう
# =========================================

# グローバル変数（関数の外で定義）
price1 = 100
price2 = 200

# 関数内でtaxをローカル変数として使う
def total():
    tax = 1.1  # ローカルスコープ
    return (price1 + price2) * tax

# 関数を呼び出して計算結果を出力
print("合計（税含む）:", total())
```
