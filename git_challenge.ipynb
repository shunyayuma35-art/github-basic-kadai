```python
# Gitのユーザー情報を設定 (ご自身の情報に置き換えてください)
!git config --global user.email "your_email@example.com"
!git config --global user.name "Your Name"

# git_challenge ディレクトリを作成し、移動
!mkdir -p git_challenge
%cd git_challenge

# Gitリポジリを初期化
!git init

# デフォルトの master ブランチを main にリネーム
!git branch -M main

# 何かファイルを追加して最初のコミットを作成（例: .gitignore）
!echo "# Ignore common Colab files" > .gitignore
!echo "__pycache__/" >> .gitignore

# hello_world.txt を作成
!echo "こんにちは、世界！" > hello_world.txt
!echo "Hello, world!" >> hello_world.txt

!git add .
!git commit -m "Initial commit: Add .gitignore and hello_world.txt"

# リモートリポジリを接続
# <ユーザー名> を実際のGitHubユーザー名に置き換えてください
!git remote add origin https://github.com/<ユーザー名>/git_challenge.git

# 初回プッシュ（mainブランチを送信）
!git push -u origin main
```

```python
# ブランチ作成
!git branch pull_request

# ブランチ切り替え
!git checkout pull_request
```
